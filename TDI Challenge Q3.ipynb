```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df_raw = pd.read_csv('KIND_2014_Final_For_Website_483300_7.csv')
```

```python
df_raw.head()
```

```python
df_raw['INCOM'] = pd.to_numeric(df_raw['INCOM'], errors='coerce')
df_raw['Total'] = pd.to_numeric(df_raw['Total'], errors='coerce')
```

```python
df_working = df_raw.groupby(['COUNTY']).sum().reset_index(drop=False).copy()
df_working['FractionIncom'] = df_working['INCOM']/df_working['Total']
df_working = df_working.sort_values(by=['FractionIncom'])
ax = df_working['FractionIncom'].plot(kind="bar")
```
